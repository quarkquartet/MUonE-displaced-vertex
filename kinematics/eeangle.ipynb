```python
import numpy as np
import ROOT as r
import random
import os
import matplotlib.pyplot as plt
# -----------------------------------------------------------------
# Import ROOT libraries for MadGraph generated rootfiles.
# -----------------------------------------------------------------
r.gROOT.ProcessLine(".include /Users/isaac/Work/MG5_aMC_v3_1_0")
r.gInterpreter.Declare(
    '#include "/Users/isaac/Work/MG5_aMC_v3_1_0/ExRootAnalysis/\
ExRootAnalysis/ExRootTreeReader.h"'
)
r.gInterpreter.Declare(
    '#include "/Users/isaac/Work/MG5_aMC_v3_1_0/\
ExRootAnalysis/ExRootAnalysis/ExRootClasses.h"'
)
r.gInterpreter.Declare(
    '#include "/Users/isaac/Work/MG5_aMC_v3_1_0/ExRootAnalysis/\
ExRootAnalysis/ExRootLHEFReader.h"'
)
r.gSystem.Load(
    "/Users/isaac/Work/MG5_aMC_v3_1_0/\
ExRootAnalysis/libExRootAnalysis.so"
)

# -----------------------------------------------------------------
# set up environment
# -----------------------------------------------------------------

os.environ["TERM"] = "linux"
random.seed(1)
```

```python
chain = r.TChain("LHEF")
chain.Reset()
chain.Add("~/Work/MUonE/MG5_events/lepton2to3vec/Events/X_mass_0.03/unweighted_events.root")
angle_list = []
E_list = []
for event in chain:
    for particle in event.Particle:
        if particle.PID == 103:
            E_list.append(particle.E)
            angle_list.append(2.0*0.03/particle.E)
            
angle_list = np.array(angle_list)
E_list = np.array(E_list)
weight_list = [1/10000.0 for i in range(10000)]
fig=plt.figure(figsize=(10,5))
fig.suptitle('mX = 0.03 GeV')
ax1=fig.add_subplot(121)
ax2=fig.add_subplot(122)
hist, bins= np.histogram(angle_list)
logbins = np.logspace(np.log10(bins[0]), np.log10(bins[-1]), len(bins))
ax1.set_xscale('log')
ax1.set_yscale('log')
ax1.set_xlabel('open angle of e+e-')
ax1.set_ylabel('normalized distribution')
ax1.hist(angle_list, bins=logbins, weights=weight_list)
hist, bins = np.histogram(E_list)
logbins = np.logspace(np.log10(bins[0]), np.log10(bins[-1]), len(bins))
ax2.set_xscale('log')
ax2.set_yscale('log')
ax2.set_xlabel('Energy of X')
ax2.hist(E_list, bins=logbins, weights=weight_list)
fig.subplots_adjust(wspace=0.3)
fig.savefig("/Users/isaac/Work/MUonE/2to3vec/e+e-angle/mX_0.03.png", dpi=300)
```
